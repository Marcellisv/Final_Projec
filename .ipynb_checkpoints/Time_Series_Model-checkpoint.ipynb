# 2. Train Model

```python
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(cali)
```

# 3. Forecast Away

```python
future = m.make_future_dataframe(periods=100,freq='D')
forecast = m.predict(future)
forecast.head()
```

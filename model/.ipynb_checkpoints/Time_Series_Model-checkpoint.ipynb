# 2. Train Model

```python
m = Prophet(interval_width=0.95, daily_seasonality=True)
m.add_regressor("add1")
m.fit(penn)
```

# 3. Forecast Away

```python
forecast = m.predict(penn.drop(columns="y"))
forecast.head()
```
